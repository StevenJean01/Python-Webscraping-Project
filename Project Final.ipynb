```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www150.statcan.gc.ca/n1/pub/36-28-0001/2022003/article/00001-eng.htm' #This is the link I will be using for the soup to webscrape the table.

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup) #Here I print the soup which is all the info I get from inspect element from the website.
```

```python
soup.find('table') #Here I use find to be able to find the table that I need.
```

```python
table = soup.find('table', class_ = 'pub-table')
```

```python
print(table) # Here I print the table.
```

```python
years = table.find_all('th') 
                                # Here I find all the lines that start with a "th" since that the lines that have all the years and then I continue to print it.
print(years)
```

```python
years_titles = [title.text.strip() for title in years]
                                                        #I then strip all the string from that line to only keep the years
print(years_titles[1:]) # Here I print it the years but start at position 1 since I do not need position 0.
```

```python
import pandas as pd

years_data = [{"Year": title} for title in years_titles[1:]]  # Exclude the first element

# Create a DataFrame
df = pd.DataFrame(years_data)

# Print the DataFrame
df

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

```

```python
import scipy.stats as stats
from sklearn.model_selection import train_test_split
# This is the Null Hypothesis for the Hypothesis test.
print('Null Hypothesis H0 = There is no correlation between work experience during studies and job placement after graduation for international students.')
print()  # Empty line
# This is the Alternative Hyphthesis for the Hypothesist test.
print('Alternative Hypothesis H1 = There is a significant correlation between work experience during studies and job placement after graduation for international students.')
print()  # Empty line
# Here I am performing a Pearson correlation test
correlation, p_value = stats.pearsonr(df6['Rate of Students Working During Studies'], df5['International Student Employment Rate'])
# Here I print all the values
print(f'Correlation: {correlation}')
print(f'P-value: {p_value}')
# Here I use a significant level of 10% to test if the correlation is statisically significant.
alpha = 0.10
if p_value < alpha:
    print('The correlation is statistically significant, reject the null hypothesis')
else:
    print('The correlation is not statistically significant, fail to reject the null hypothesis')
```

```python
from sklearn import metrics 
# Here I divid the percentages by 100 so that it is easier to work with.
df6 = df6/100
df5 = df5/100
#df6['Rate of Students Working During Studies'] = df5['International Student Employment Rate'] / 100
#Here I set my x and y variables.
X1 = df6[['Rate of Students Working During Studies']]
y1 = df5['International Student Employment Rate']
# Here I split the data into training and testing sets.
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
# Here I create a linear regression model.
model = LinearRegression()
# Here I train the model.
model.fit(X1_train, y1_train)
# Here I make predictions on the test set.
y_pred = model.predict(X1_test)
# Here I evaluate the results.
print('Mean Absolute Error:', metrics.mean_absolute_error(y1_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y1_test, y_pred))
print('Root Mean Squared Error:', metrics.mean_squared_error(y1_test, y_pred, squared=False))

rounded_coefficient = round(metrics.mean_absolute_error(y1_test, y_pred), 4)  # Adjust the number of decimal places as needed
print(f'Coefficient: {rounded_coefficient}')

# Here I plot the regression line
plt.scatter(X1_test, y1_test, color='black')
plt.plot(X1_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Students Whom Work While Studying')
plt.ylabel('Employed Students After Graduation')
plt.title('Linear Regression: International Students Working During Studies vs Employed After Graduation')
plt.show()
```
